```python
import sys
import os
import re
import json
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
topdir = "/vol/dhcp-derived-data/volumetric-atlases/groupwise"
```

```python
tissue_names = ["CSF", "cGM", "WM", "BG", "Ventricles", "CB", "dGM", "BS", "Hippo+Amygd"]
all_tissues = ["BG", "WM", "cGM", "dGM", "CSF", "Ventricles", "BS+CB"]
tissues = ["WM", "cGM", "dGM"]

def savefig(fig, fname, infix="pairwise-"):
    fig.savefig("brain-atlas-" + infix + "paramsearch-" + fname + ".pdf", dpi=1200, bbox_inches="tight")

def measure_title(measure):
    if measure == "gini":
        return "Gini coefficient"
    if measure == "sdev":
        return "StDev"
    if measure == "dice":
        return "DSC"
    if measure == "lc_tissues":
        return "Label consistency"
    if measure == "lc":
        return "Label consistency"
    return measure

def tissue_structures(name):
    lname = name.lower()
    if lname in ("csf", "cerebrospinal fluid"):
        return [83]
    if lname in ("cgm", "gm", "cortical grey matter", "cortical gray matter", "grey matter", "gray matter"):
        return list(range(5, 17)) + list(range(20, 40))
    if lname in ("wm", "white matter"):
        return list(range(51, 83))
    if lname in ("bg", "background"):
        return [84]
    if lname in ("vent", "vents", "ventricle", "ventricles", "lateral ventricle", "lateral ventricles"):
        return [49, 50]
    if lname in ("bs+cb", "brainstem and cerebellum"):
        return [17, 18, 19]
    if lname in ("dgm", "deeg grey matter", "deep gray matter"):
        return [1, 2, 3, 4] + list(range(40, 49)) + list(range(85, 88))
    raise ValueError("Unknown tissue class: {0}".format(name))
```

```python
ages = (36, 40, 44)

def wtostr(w):
    return format(w, ".0e").replace("e-0", "e-").replace("e+0", "e+").replace("0e+0", "0e-0")
    
def testdir(sim, ds, be, lj, step, bg=False):
    if bg:
        fmt = "sim_{sim}-ds_{ds:.1f}-be_{be}-lj_{lj}-bg_y/cache/i{step:02d}"
    else:
        fmt = "sim_{sim}-ds_{ds:.1f}-be_{be}-lj_{lj}/cache/i{step:02d}"
    subdir = fmt.format(sim=sim.lower(), ds=ds, be=wtostr(be), lj=wtostr(lj), step=step)
    return os.path.join(topdir, "tests", subdir)
```

```python
table = None
for sim in ["NMI"]:
    for ds in [2.0, 2.5]:
        for be in [1e-4, 5e-3, 1e-3, 1e-2]:
            for lj in [0, 1e-5, 1e-4, 1e-3]:
                for step in [1, 2, 3]:
                    outdir = testdir(sim=sim, ds=ds, be=be, lj=lj, step=step, bg=False)
                    if not os.path.exists(outdir):
                        raise ValueError("MISSING: " + outdir)
                    for t in [36, 40, 44]:
                        df = pd.read_csv(os.path.join(outdir, "qc-measures", "t{t:.0f}.csv".format(t=t)))
                        df = df.assign(sim=sim, ds=ds, be=be, lj=lj, i=step, t=t)
                        if table is None:
                            table = df
                        else:
                            table = pd.concat([table, df])
```

```python
table.rename(columns={
    "tissues/grad_2": "cGM gradient magnitude",
    "structures/label-consistency": "Label consistency",
    "t2w/grad": "T2-w gradient magnitude",
    "t2w/gini": "Gini coefficient",
    "roi": "ROI",
    "sim": "Similarity",
    "ds": "Spacing",
    "be": "BE weight",
    "lj": "JAC weight",
    "i": "Iterations",
    "t": "Age"
}, inplace=True)
table.head()
```

```python
sns.factorplot(x="BE weight", y="Label consistency", hue="Iterations", col="JAC weight", row="Spacing", data=table)
```

# Term atlas parameter search

```python
def read_regcfg(config, step):
    with open(os.path.join(topdir, "config", "term", config + ".json"), "rt") as f:
        cfgs = json.load(f)
    cfgs = cfgs.get("registration", {}).get("config", {})
    if not isinstance(cfgs, list):
        cfgs = [cfgs]
    cfg = {}
    for i in range(step):
        cfg.update(cfgs[i])
    return cfg

def read_params(config, step):
    cfg = read_regcfg(config, step)
    params = {
        "Similarity": cfg.get("measures", {}).get("t2w", "NMI"),
        "Spacing": cfg.get("spacing", 2.0),
        "BE weight": cfg.get("bending", 5e-3),
        "JAC weight": cfg.get("jacobian", 1e-5),
        "Model": cfg.get("model", "SVFFD"),
        "Energy": cfg.get("energy", "Symmetric"),
        "Integration method": cfg.get("ffdim", "SS"),
        "No. of integration steps": cfg.get("intsteps", 64),
        "No. of BCH terms": cfg.get("bchterms", 4),
        "Use Jacobian in Lie bracket": cfg.get("liederiv", False),
        "Exclude background": cfg.get("padding", True),
        "Foreground": cfg.get("foreground", "Overlap"),
        "Interpolation": cfg.get("interpolation", "Linear with padding")
    }
    if isinstance(params["Spacing"], list):
        params["Spacing"] = params["Spacing"][0]
    return params

def read_measures(config, step, t=40.0):    
    df = pd.read_csv(os.path.join(topdir, "output", "term", config, "i{0:02d}".format(step), "qc-measures", "t{t:.0f}.csv".format(t=t)))
    del df["t2w/mean"]
    df.rename(columns={
        "roi": "ROI",
        "structures/label-consistency": "Label consistency",
        "t2w/gini": "Gini coefficient",
        "t2w/sdev": "Standard deviation"
    }, inplace=True)
    params = read_params(config, step)
    params["ConfigId"] = config
    params["Iteration"] = step
    df = df.assign(**params)
    df.insert(df.columns.get_loc("ROI"), "Tissue", None)
    for l in range(1, 10):
        df.loc[df["ROI"].isin(["tissue={}".format(l), "roi={}".format(l)]),"Tissue"] = tissue_names[l - 1]
    for tissue in all_tissues:
        df.loc[df["ROI"].isin(["structure={:02d}".format(l) for l in tissue_structures(tissue)]),"Tissue"] = tissue
    return df
```

```python
table = None
sims = ["nmi", "ncc"]
for sim in sims:
    for c in range(40, 49):
        config = "c{0:02d}".format(c)
        if sim == "ncc":
            config = config + "-ncc"
        df = read_measures(config=config, step=5)
        if table is None:
            table = df
        else:
            table = pd.concat([table, df])
structure_measures = table[table["ROI"].isin(["structure={}".format(l) for l in range(1, 88)])]
tissue_measures = table[table["ROI"].isin(["tissue={}".format(l) for l in range(1, 10)])]
global_measures = table[table["ROI"].isin(["roi={}".format(l) for l in range(1, 10)])]
```
